This code was used to obtain the LaTeX code for the table in the reports.

```python
import re
import pandas as pd
```

## CP

```python
txtfile = '../CP/out/log.txt'
with open(txtfile, 'r') as f:
    lines = [l.rstrip() for l in f.readlines()]
r1 = re.compile(r".* (Final|optimal) solution .*")
r2 = re.compile(r"Solving problem *")
nonoptimal_reg = re.compile(r"Couldn't")
time_reg = re.compile(r"\d*\b\.\d*\b")
solutions_times = [(float(time_reg.search(l).group(0)), 
                    False if nonoptimal_reg.match(l) else True) 
                   for l in list(filter(r1.match, lines))]
instances = [int(l.split('ins-')[1].split('.txt')[0]) 
             for l in list(filter(r2.match, lines))]

instance_times = {instances[i]: solutions_times[i] for i in range(len(instances))}
instance_times
```

```python
results = pd.DataFrame(instance_times, index=['Solving Time', 'Solved']).T
results.sort_index()
```

```python
print(results.to_latex())
```

Join

```python
joined_table = results.join(results_rot, rsuffix=' (rotated)').sort_index()
joined_table
```

```python
print(joined_table.to_latex())
```

## SAT

```python
txtfile = '../SAT/out/log.txt'
rot_txtfile = '../SAT/out/log-rot.txt'
with open(txtfile, 'r') as f:
    lines = [l.rstrip() for l in f.readlines()]
with open(rot_txtfile, 'r') as f:
    lines_rot = [l.rstrip() for l in f.readlines()]

r1 = re.compile(r".* (Final|optimal) solution .*")
r2 = re.compile(r"Solving problem *")
nonoptimal_reg = re.compile(r"Couldn't")
time_reg = re.compile(r"\d*\b\.\d*\b")

solutions_times = [(float(time_reg.search(l).group(0)), 
                    False if nonoptimal_reg.match(l) else True) 
                   for l in list(filter(r1.match, lines))]
instances = [int(l.split('ins-')[1].split('.txt')[0]) 
             for l in list(filter(r2.match, lines))]

solutions_times_rot = [(float(time_reg.search(l).group(0)), 
                        False if nonoptimal_reg.match(l) else True) 
                        for l in list(filter(r1.match, lines_rot))]
instances_rot = [int(l.split('ins-')[1].split('.txt')[0]) 
                for l in list(filter(r2.match, lines_rot))]

instance_times = {instances[i]: solutions_times[i] for i in range(len(instances))}
instance_times_rot = {instances_rot[i]: solutions_times_rot[i] for i in range(len(instances_rot))}

results = pd.DataFrame(instance_times, index=['Solving Time', 'Solved']).T
results_rot = pd.DataFrame(instance_times_rot, index=['Solving Time', 'Solved']).T
joined_table = results.join(results_rot, rsuffix=' (rotated)').sort_index()

print(joined_table.to_latex())
```

```python
txtfile = '../SMT/out/log.txt'
with open(txtfile, 'r') as f:
    lines = [l.rstrip() for l in f.readlines()]

r1 = re.compile(r".* (Final|optimal) solution .*")
r2 = re.compile(r"Solving problem *")
nonoptimal_reg = re.compile(r"Couldn't")
time_reg = re.compile(r"\d*\b\.\d*\b")

def replace_with_geq300(time):
    if float(time) >= 300:
        return "geq 300"
    else:
        return time

solutions_times = [(float(time_reg.search(l).group(0)), 
                    False if nonoptimal_reg.match(l) else True) 
                   for l in list(filter(r1.match, lines))]
instances = [int(l.split('ins-')[1].split('.txt')[0]) 
             for l in list(filter(r2.match, lines))]

instance_times = {instances[i]: solutions_times[i] for i in range(len(instances))}

results = pd.DataFrame(instance_times, index=['Solving Time', 'Solved']).T.sort_index()
results['Solving Time'] = results['Solving Time'].apply(replace_with_geq300)

print(results.to_latex())
```

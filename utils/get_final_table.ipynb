This code was used to obtain the LaTeX code for the table in the CP report.

```python
import re
import pandas as pd
```

```python
txtfile = '../CP/out/log.txt'
with open(txtfile, 'r') as f:
    lines = [l.rstrip() for l in f.readlines()]
r1 = re.compile(r".* (Final|optimal) solution .*")
r2 = re.compile(r"Solving problem *")
nonoptimal_reg = re.compile(r"Couldn't")
time_reg = re.compile(r"\d*\b\.\d*\b")
solutions_times = [(float(time_reg.search(l).group(0)), 
                    False if nonoptimal_reg.match(l) else True) 
                   for l in list(filter(r1.match, lines))]
instances = [int(l.split('ins-')[1].split('.txt')[0]) 
             for l in list(filter(r2.match, lines))]

instance_times = {instances[i]: solutions_times[i] for i in range(len(instances))}
instance_times
```

```python
results = pd.DataFrame(instance_times, index=['Solving Time', 'Solved']).T
results.sort_index()
```

```python
print(results.to_latex())
```

Repeat for rotated

```python
txtfile = '../CP/out/log-rot.txt'
with open(txtfile, 'r') as f:
    lines = [l.rstrip() for l in f.readlines()]

solutions_times = [(float(time_reg.search(l).group(0)), 
                    False if nonoptimal_reg.match(l) else True) 
                   for l in list(filter(r1.match, lines))]
instances = [int(l.split('ins-')[1].split('.txt')[0]) 
             for l in list(filter(r2.match, lines))]

instance_times = {instances[i]: solutions_times[i] for i in range(len(instances))}
instance_times
```

```python
results_rot = pd.DataFrame(instance_times, index=['Solving Time', 'Solved']).T
results_rot.sort_index()
```

Join

```python
joined_table = results.join(results_rot, rsuffix=' (rotated)').sort_index()
joined_table
```

```python
print(joined_table.to_latex())
```
